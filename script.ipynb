```python
# Import nessecary libraries
import pandas as pd
import ast
import os
import plotly.express as px
```

```python
# Load the data into a DataFrame
df_employee = pd.read_csv('data/employee.csv')

# Inspect the data
df_employee.info()
df_employee.head(10)
```

```python
df_employee.describe()
```

```python
def seniority_to_months(seniority):
    years, months = map(int, seniority.replace('years', ' ').replace('months', '').split(','))
    return years * 12 + months
```

```python
df_employee['seniority (months)'] = df_employee['Seniority'].apply(seniority_to_months)
df_employee = df_employee.rename(columns={'Seniority': 'seniority', 'Salary': 'salary'})
df_employee.head()
```

```python
# Delete the 'Unnamed: 0' column
df_employee_handle = df_employee_handle.drop('Unnamed: 0', axis=1)

# convert the 'weekend_date' column to datetime format
df_employee_handle['weekend_date'] = pd.to_datetime(df_employee_handle['weekend_date'])

#delete the null values in the 'total_handle_time_sec' column
df_employee_handle = df_employee_handle.dropna(subset=['total_handle_time_sec'])
df_employee_handle.head()
```

```python
# Load the data into a DataFrame
df_employee_productivity = pd.read_csv('data/employee_productivity.csv')

# Inspect the data
df_employee_productivity.info()
df_employee_productivity.head(10)
```

```python
df_employee_productivity.describe()
```

```python
# convert reporting_date to datetime format
df_employee_productivity['reporting_date'] = pd.to_datetime(df_employee_productivity['reporting_date'])

# Parse the 'status_map' column
df_employee_productivity['status_map'] = df_employee_productivity['status_map'].apply(ast.literal_eval)
status_df = df_employee_productivity['status_map'].apply(pd.Series)

# Insert the parsed columns at a specified position
insert_position = 6 

for col in status_df.columns[::-1]:  # Reverse the column order for correct insertion
    df_employee_productivity.insert(insert_position, col, status_df[col])

# Drop the original 'status_map' column
df_employee_productivity.drop(columns=['Unnamed: 0'], inplace=True)
df_employee_productivity.drop(columns=['status_map'], inplace=True)

df_employee_productivity.head()
```

```python
# Verify duplicates if exists
df_employee_productivity[df_employee_productivity.duplicated()].shape
```

# Extract Facts and Dimensions

```python
# Get copy from the original dataframes
employee_handle = df_employee_handle.copy()
employee = df_employee.copy()
employee_productivity = df_employee_productivity.copy()
```

### Create Dimension Tables

```python
# Work City Dimension
work_city_dim = pd.DataFrame(employee['work_city'].unique(), columns=['title'])
work_city_dim['ID'] = range(0, len(work_city_dim))
id_col = work_city_dim.pop('ID')  
work_city_dim.insert(0, 'ID', id_col)
```

```python
# Employee Dimension
employee_dim = employee[['employee_id', 'seniority']].drop_duplicates()
```

```python
# Date Dimension (from both reporting_date and weekend_date)
def create_date_dim(dates):
    dates = pd.to_datetime(dates).drop_duplicates().reset_index(drop=True)
    dates = dates.sort_values().reset_index(drop=True)
    date_dim = pd.DataFrame({
        'date': dates,
        'year': dates.dt.year,
        'month': dates.dt.month,
        'day': dates.dt.day
    })
    return date_dim

reporting_dates = employee_productivity['reporting_date']
weekend_dates = employee_handle['weekend_date']
all_dates = pd.concat([reporting_dates, weekend_dates]).drop_duplicates()
date_dim = create_date_dim(all_dates)

date_dim['ID'] = range(0, len(date_dim))
id_col = date_dim.pop('ID')  
date_dim.insert(0, 'ID', id_col)
```

```python
# Business Unit Dimension
business_unit_dim = pd.DataFrame(employee_handle['business_unit'].unique(), columns=['title'])
business_unit_dim['ID'] = range(0, len(business_unit_dim))
id_col = business_unit_dim.pop('ID')  
business_unit_dim.insert(0, 'ID', id_col)
```

```python
# Market Dimension
market_dim = pd.DataFrame(employee_handle['market'].unique(), columns=['title'])
market_dim['ID'] = range(0, len(market_dim))
id_col = market_dim.pop('ID')  
market_dim.insert(0, 'ID', id_col)
```

```python
# Event Dimension
event_dim = pd.DataFrame(employee_handle['event'].unique(), columns=['title'])
event_dim['ID'] = range(0, len(event_dim))
id_col = event_dim.pop('ID')  
event_dim.insert(0, 'ID', id_col)
```

### Create Fact Tables

```python
# Employee Details Fact
employee_details_fact = employee[['employee_id', 'salary', 'seniority (months)', 'work_city']]
employee_details_fact = employee_details_fact.rename(columns={'seniority (months)': 'seniority_months'})
employee_details_fact = employee_details_fact.merge(work_city_dim, left_on='work_city', right_on='title')
employee_details_fact = employee_details_fact.rename(columns={'ID': 'work_city_id'})
employee_details_fact = employee_details_fact[['employee_id', 'salary', 'seniority_months', 'work_city_id']]
employee_details_fact['ID'] = range(0, len(employee_details_fact))
id_col = employee_details_fact.pop('ID')  
employee_details_fact.insert(0, 'ID', id_col)
```

```python
# Employee Productivity Fact
employee_productivity_fact = employee_productivity.merge(work_city_dim, left_on='work_city', right_on='title', suffixes=('_employee', '_city'))
employee_productivity_fact = employee_productivity_fact.merge(date_dim, left_on='reporting_date', right_on='date', suffixes=('_city', '_date'))
employee_productivity_fact = employee_productivity_fact.merge(business_unit_dim, left_on='business_unit', right_on='title', suffixes=('_date', '_business_unit'))
employee_productivity_fact = employee_productivity_fact.merge(market_dim, left_on='market', right_on='title', suffixes=('_business_unit', '_market'))
employee_productivity_fact = employee_productivity_fact.rename(columns={
    'ID_city': 'work_city_id', 
    'ID_date': 'reporting_date_id', 
    'ID_business_unit': 'business_unit_id', 
    'ID_market': 'market_id'
})
employee_productivity_fact = employee_productivity_fact[[
    'employee_id', 'unavailable_sec', 'available_sec', 'meal_sec', 'coaching_sec', 'training_sec', 'onboarding_sec', 
    'time_in_billable_statuses_sec', 'onboarding_status', 'work_city_id', 'reporting_date_id', 'business_unit_id', 'market_id'
]]
employee_productivity_fact['ID'] = range(0, len(employee_productivity_fact))
id_col = employee_productivity_fact.pop('ID')  
employee_productivity_fact.insert(0, 'ID', id_col)
```

```python
# Employee Handle Fact
employee_handle_fact = employee_handle.merge(work_city_dim, left_on='work_city', right_on='title', suffixes=('_employee', '_city'))
employee_handle_fact = employee_handle_fact.merge(date_dim, left_on='weekend_date', right_on='date', suffixes=('_city', '_weekend_date'))
employee_handle_fact = employee_handle_fact.merge(business_unit_dim, left_on='business_unit', right_on='title', suffixes=('_weekend_date', '_business_unit'))
employee_handle_fact = employee_handle_fact.merge(market_dim, left_on='market', right_on='title', suffixes=('_business_unit', '_market'))
employee_handle_fact = employee_handle_fact.merge(event_dim, left_on='event', right_on='title', suffixes=('_market', '_event'))
employee_handle_fact = employee_handle_fact.rename(columns={
    'ID_city': 'work_city_id', 
    'ID_weekend_date': 'weekend_date_id', 
    'ID_business_unit': 'business_unit_id', 
    'ID_market': 'market_id',
    'ID': 'event_id'
})
employee_handle_fact = employee_handle_fact[[
    'employee_id', 'total_handle_time_sec', 'aht_denom', 'work_city_id', 'weekend_date_id', 'business_unit_id', 'market_id', 'event_id'
]]

employee_handle_fact['ID'] = range(0, len(employee_handle_fact))
id_col = employee_handle_fact.pop('ID')  
employee_handle_fact.insert(0, 'ID', id_col)
```

### Export data to output files

```python
# Create the 'output' directory if it doesn't exist
os.makedirs('output', exist_ok=True)

# Save the new dataframes to CSV files
employee_dim.to_csv('output/Dim_Employee.csv', index=False)
work_city_dim.to_csv('output/Dim_Work_city.csv', index=False)
date_dim.to_csv('output/Dim_Date.csv', index=False)
business_unit_dim.to_csv('output/Dim_Business_unit.csv', index=False)
market_dim.to_csv('output/Dim_Market.csv', index=False)
event_dim.to_csv('output/Dim_Event.csv', index=False)
employee_details_fact.to_csv('output/Fact_Employee_details.csv', index=False)
employee_productivity_fact.to_csv('output/Fact_Employee_productivity.csv', index=False)
employee_handle_fact.to_csv('output/Fact_Employee_handle.csv', index=False)
```

### Check for Correlation

```python
#check for correlation in employee details fact between seniority and salary
correlation_matrix = employee_details_fact[['seniority_months', 'salary']].corr()

# Modify Correlation Matrix Column Names (Before Plotting):
correlation_matrix = correlation_matrix.rename(columns={'seniority_months': 'Seniority per months',
                                                      'salary': 'Salary'})

# Create the heatmap using plotly.express
fig = px.imshow(correlation_matrix,
                x=correlation_matrix.columns,
                y=correlation_matrix.columns,
                color_continuous_scale='RdBu',  # Use a diverging color scale
                zmin=-1, zmax=1,                # Set color range for correlation (-1 to 1)
                labels=dict(x="", y="", color="Correlation"),
                text_auto=True,
                width=800, 
                height=500,
                title="Correlation Heatmap: seniority per months vs. Salary"
               )
fig.show()
```
